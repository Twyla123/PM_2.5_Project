```python
data1201['ozone_Measurement']=data1201['ozone_Measurement'].apply(lambda x: 0 if x < 0 else x)
#Converting all negative values of ozone to 0
#I chose to do this because the negative values that ozone does have are
#very close to 0
```

```python
#Outliers defined here using IQR
#Takes all outliers and replaces them with the upper or lower bound respective to where they lie outside of the IQR
#Data is the dataset one wishes to clean
#Column name refers to the feature
#Returns a dataset that has been cleaned for outliers
def replace_outliers(data, column_name):
    q1 = np.percentile(data[column_name], 25)
    q3 = np.percentile(data[column_name], 75)
    IQR = q3 - q1
    lower_bound = q1 - 1.5 * IQR
    upper_bound = q3 + 1.5 * IQR

    data[column_name] = np.where(data[column_name] < lower_bound, lower_bound, data[column_name])
    data[column_name] = np.where(data[column_name] > upper_bound, upper_bound, data[column_name])
    
    return data
```

```python
replace_outliers(data1103, 'ozone_Measurement')
replace_outliers(data1201, 'ozone_Measurement')
```

HW Summary:

All negative values have been replaced with zero. This choice has been made due to all negative values being extremely close to 0, implying some rounding error or error in collecting the data.

All outliers have been identified using the IQR. Outliers have been replaced by upper and lower bounds defined by the IQR. I implemented this method because I felt that I knew how to do it confidently and don't really know any other way.
