```python
#Calculates upper and lower bounds based on IQR
#Replaces outliers in the data
#Returns a cleaned dataset
def outlier_processing(data, column_name):
    q1 = np.percentile(data[column_name], 25)
    q3 = np.percentile(data[column_name], 75)
    IQR = q3 - q1
    lower_bound = q1 - 1.5 * IQR
    upper_bound = q3 + 1.5 * IQR

    #Checks if a value is an outlier. If it is, it works to replace it, prioritizing previous hour, then next hour, then previous year, then next year
    #If all of those values are outliers, it replaces it with an upper or lower bound
    #Returns the new value for that row that is no longer an outlier
    def replace_outlier(row):
        nonlocal data, column_name
        index = row.name
        value = row[column_name]
        
        current_time = row['datetime']
        
        if value > upper_bound or value < lower_bound:
            previous_hour = current_time - pd.Timedelta(hours=1)
            if previous_hour in data.index and data.loc[previous_hour, column_name] <= upper_bound and data.loc[previous_hour, column_name] >= lower_bound:
                row[column_name] = data.loc[previous_hour, column_name]
            next_hour = current_time + pd.Timedelta(hours=1)
            if next_hour in data.index and data.loc[next_hour, column_name] <= upper_bound and data.loc[next_hour, column_name] >= lower_bound:
                row[column_name] = data.loc[next_hour, column_name]
            previous_year = current_time - pd.DateOffset(years=1)
            if previous_year in data.index and data.loc[previous_year, column_name] <= upper_bound and data.loc[previous_year, column_name] >= lower_bound:
                row[column_name] = data.loc[previous_year, column_name]
            next_year = current_time + pd.DateOffset(years=1)
            if next_year in data.index and data.loc[next_year, column_name] <= upper_bound and data.loc[next_year, column_name] >= lower_bound:
                row[column_name] = data.loc[next_year, column_name]
            else:
                if value > upper_bound:
                    row[column_name] = upper_bound
                elif value < lower_bound:
                    row[column_name] = lower_bound
        return row

    data = data.apply(replace_outlier, axis=1)
    return data
```

```python
data1103_clean = outlier_processing(data1103, 'ozone_Measurement')
data1201_clean = outlier_processing(data1201, 'ozone_Measurement')
```

```python
def detect_outliers_iqr(df, column_name):
   
    Q1 = df[column_name].quantile(0.25)
    Q3 = df[column_name].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    
    outliers = df[(df[column_name] < lower_bound) | (df[column_name] > upper_bound)]

    num_outliers = outliers.shape[0]
    
    return num_outliers, outliers


num_outliers_1103, outliers_1103 = detect_outliers_iqr(data1103, 'ozone_Measurement')
print(f"Number of outliers detected: {num_outliers_1103}")

outliers_1103
```

```python
num_outliers_1201, outliers_1201 = detect_outliers_iqr(data1201, 'ozone_Measurement')
print(f"Number of outliers detected: {num_outliers_1201}")

outliers_1201
```
